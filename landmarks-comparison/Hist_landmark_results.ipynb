```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('co-astro-landmark-analysis')

sns.set(style="whitegrid")  

fig = plt.figure(figsize = (12,12))
fig.suptitle("Approximation Error \n co-astro", fontsize=24, x = 0.5, y=0.97)

plt.subplot(3, 2, 1)

plt.bar(data[0:4]["Number of Landmarks"], data[0:4]["Approximation Error"], color ='orange', width = 8)
plt.title("basic", fontsize=14)
plt.ylabel("random")

plt.subplot(3, 2, 2)
plt.bar(data[12:16]["Number of Landmarks"], data[12:16]["Approximation Error"], color ='orange', width = 8)
plt.title("LCA", fontsize=14)

plt.subplot(3, 2, 3)
plt.bar(data[4:8]["Number of Landmarks"], data[4:8]["Approximation Error"], color ='orange', width = 8)
plt.ylabel("degree")

plt.subplot(3, 2, 4)
plt.bar(data[16:20]["Number of Landmarks"], data[16:20]["Approximation Error"], color ='orange', width = 8)

plt.subplot(3, 2, 5)
plt.bar(data[8:12]["Number of Landmarks"], data[8:12]["Approximation Error"], color ='orange', width = 8)
plt.ylabel("coverage")

plt.subplot(3, 2, 6)
plt.bar(data[20:24]["Number of Landmarks"], data[20:24]["Approximation Error"], color ='orange', width = 8)
plt.show()
```
