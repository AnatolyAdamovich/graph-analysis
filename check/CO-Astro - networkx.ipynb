```python
import networkx as nx
```

```python
Graph = nx.Graph()

with open('../datasets/CA-AstroPh.txt', mode='r') as f:
    for line in f:
        u, v = line.split()
        Graph.add_edge(u, v)
```

```python
print('Число вершин:', Graph.number_of_nodes())
print('Число рёбер:', Graph.number_of_edges())
```

```python
print('Число компонент связности - ', nx.number_connected_components(Graph))
components = list(nx.connected_components(Graph))
```

```python
largest_comp = Graph.subgraph(max(components, key = lambda c: len(c)))
print('Число вершин в наибольшей компоненте:', largest_comp.number_of_nodes())
print('Число рёбер в наибольшей компоненте:', largest_comp.number_of_edges())
```
